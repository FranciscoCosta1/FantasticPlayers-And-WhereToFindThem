# Football Players ID Match between Transfer Markt DB & FIFA 19/20 DB

## Import libraries

```python
import time
import pandas as pd
import Levenshtein as Lev
```

## Import CSV files

```python
Transfers_Market_filepath = './Input Files/TOP Players.csv'
FIFA_Player_Stats_filepath = './Input Files/players _ fifa20.csv'

df_playersTM = pd.read_csv(Transfers_Market_filepath, sep = ';').loc[:,['PlayerID', 'Player', 'Age', 'Team', 'Position']]
df_playersFIFA = pd.read_csv(FIFA_Player_Stats_filepath).loc[:,['sofifa_id', 'short_name', 'age', 'club']]
```

## Convert DataFrame columns to a list of tuples (Player ID, Player Name, Current Team)

```python
TM = list(df_playersTM.itertuples(index = False, name = None))
```

```python
FIFA = list(df_playersFIFA.itertuples(index = False, name = None))
```

## String similarity (using Levenshtein algorithm) between the two lists

```python
match_list = []
start_time = time.time()

for tm_id, tm_player, tm_age, tm_team, tm_position in TM:
    
    dmax = 1000
    
    for fifa_id, fifa_player, fifa_age, fifa_team in FIFA:
        
        name_dist = Lev.distance(tm_player, fifa_player)
        name_team_dist = Lev.distance(tm_player + ' - ' + tm_team, fifa_player + ' - ' + fifa_team)
        avg_dist = (name_dist + name_team_dist)/2
        age_dif = abs(tm_age - fifa_age)

        if avg_dist < dmax:
            match_row = [tm_id, tm_player, tm_age, tm_team, tm_position, fifa_id, fifa_player, fifa_age, fifa_team, 
                         name_dist, name_team_dist, avg_dist, age_dif]
            dmax = avg_dist
            
    match_list.append(match_row)

df = pd.DataFrame(match_list)    
print('Process completed in:', time.time() - start_time, 'seconds.')
```
